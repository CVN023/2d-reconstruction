# 2D image reconstruction

We tackle the ill-posed inverse problem which consists in finding a source $d \in \mathbb{R}^n$ such that $p = Ad$ is verified, where $p \in \mathbb{R}^m$ is an observed projection and $A \in \mathbb{R}^{m \times n}$ represents the projection operator of the imaging system.

Here, we consider an underdetermined problem with $m<n$ ($A$ is a *fat matrix*) which consequently has an infinite number of solutions.

In what follows, $A$ is supposed to be full-rank ($rank(A)=m$).

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_context('poster')
sns.set_color_codes()
plot_kwds = {'alpha' : 0.25, 's' : 80, 'linewidths':0}

from PIL import Image
import scipy as sp
import pylab
```

Import image:

```python
cortex  = Image.open('cortex.bmp')
cortex = np.asarray(cortex)[:,:,0]
# Downsampling
cortex = np.asarray(cortex)[::6,::6]

plt.imshow(cortex)
```

Image size:

```python
n = cortex.shape[0]
print(cortex.shape)
```

Projection size:

```python
m = int(3*n/4)
print(m)
```

Projection matrix:

```python
np.random.seed(5)
A = 20 + np.random.randn(m**2,n**2)#5*np.random.randn(m**2,n**2)
```

Construct projection:

```python
p = np.matmul(A, cortex.flatten())
```

### Minimal norm solution

We first tackle the minimal norm problem. This problem can be written as:
$$
\left\{
    \begin{array}{ll}
        \text{argmin}_d \frac{1}{2} ||d||^2 \\
        \text{s.t. } Ad = p
    \end{array}
\right.
$$

It can be easily shown (see http://www.math.usm.edu/lambers/mat419/lecture15.pdf) that if $\hat{d}$ is solution of the above problem, then $\hat{d}$ is in the range of $A^T$:
there exists $w \in \mathbb{R}^m$ such that $\hat{d} = A^T w$.

But as $\hat{d}$ must verify $A\hat{d}=p$, we must have $A A^T w = p$, i.e. $w = (A A^T)^{-1}p$ since we assumed that $A$ is full-rank. Thus we simply have: $\hat{d} = A^T (A A^T)^{-1}p$.

```python
inverse = np.linalg.inv(np.matmul(A, A.T))
```

```python
approx = np.matmul(A.T, np.matmul(inverse, p))
```

```python
approx.shape
```

```python
plt.imshow(approx.reshape(n,n))
```

Evaluation metric (should be maximized):

```python
def snr(x, y):
    return 20 * np.log10(pylab.norm(x) / pylab.norm(x - y))
```

```python
snr(cortex.flatten(), approx)
```

# Total Variation Regularization

Here the problem we try to solve is the following:

$$ \text{argmin}_d J(d) \qquad \text{with} \qquad J(d) = \frac{1}{2} ||Ad - p||^2 + \lambda TV(d)\\ $$


where $TV(d) = \sum_x ||\nabla_x d(x)||$ is the measure of the total variation of the gradient on the reconstructed image.

$TV$ is not a smooth function of the image $d$: in order to be able to differentiate $TV$ we consider a smoothed version $TV_{\epsilon}(d) = \sum_x \sqrt{||\nabla_x d(x)||^2 + \epsilon^2}$ where $\epsilon$ is a small real $>0$.


The above problem with the smoothed regulariation $TV_{\epsilon}$ is a convex problem that can be solved with gradient descent. Here, the gradient of the cost function will be:

$$ \nabla_d J(d) = A^T(Ad-p) - \lambda div( \frac{\nabla_x d}{\sqrt{||\nabla_x d(x)||^2 + \epsilon^2}} )$$

where $\nabla_x$ is the gradient operator in 2D and $div$ is the divergence operator in 2D, both discretized on a pixel grid (see https://mathematical-tours.github.io/book-sources/chapters-pdf/inverse-problems.pdf for details)

Definition of gradient and divergence (on 2D images with finite approximation and periodicity hypothesis):

```python
def grad(f):
    S = f.shape;
    s0 = np.concatenate( (np.arange(1,S[0]),[0]) )
    s1 = np.concatenate( (np.arange(1,S[1]),[0]) )
    g = np.dstack( (f[s0,:] - f, f[:,s1] - f))
    return g
```

```python
def div(g):
    S = g.shape;
    s0 = np.concatenate( ([S[0]-1], np.arange(0,S[0]-1)) )
    s1 = np.concatenate( ([S[1]-1], np.arange(0,S[1]-1)) )
    f = (g[:,:,0] - g[s0,:,0]) + (g[:,:,1]-g[:,s1,1])
    return f
```

Variations in the original image:

```python
plt.imshow( np.sqrt(np.sum(grad(cortex)**2, axis=2)) )
```

Initialisation in the image domain:

```python
dTV = np.matmul(A.T, p)
```

Inital energy term of reconstruction:

```python
1/2*np.linalg.norm( (np.matmul(A, dTV) - p).flatten())**2 
```

* Smoothing parameter $\epsilon$
* Regularization parameter $\lambda$

```python
epsilon = 1e-5#7
Lambda = 8.6e8 #2.0e9
```

Gradient of total variation:

```python
Gr = grad(dTV.reshape(n,n))
Den = np.sqrt(epsilon**2 + np.sum(Gr**2, axis=2))
G = -div(Gr / np.repeat(np.expand_dims(Den, axis=-1), 2, axis=-1) )
```

Smoothed TV norm:

```python
tv = np.sum(Den.flatten())
print('Smoothed TV norm of the image: ' + str(round(tv,2)) )
```

```python
np.min(cortex)
```

```python
np.max(cortex)
```

Gradient descent with adaptative learning rate (steepest descent for the reconstruction objective, not for the total objective...):

```python
niter = 50#700

dTV = np.matmul(A.T, p)
E = np.zeros((niter,1))
rec = np.zeros((niter,1))
TV = np.zeros((niter,1))
for i in range(niter):
    # compute the gradient of the euclidean reconstruction error
    e = np.matmul(A, dTV) - p
    grad_rec = np.matmul(A.T, e)
    # compute the gradient of the smoothed TV functional.
    Gr = grad(dTV.reshape(n,n))
    Den = np.sqrt(epsilon**2 + np.sum(Gr**2, axis=2))
    G = -div(Gr / np.repeat(np.expand_dims(Den, axis=-1), 2, axis=-1) )
    # add the two gradients to form the update vector r
    r = grad_rec + Lambda*G.flatten()
    # compute step size according to energy reconstruction values
    #mu = np.matmul(r, grad_rec) / np.linalg.norm(np.matmul(A, r))**2
    tv_term = np.matmul(r, -div( (Gr / np.repeat(np.expand_dims(Den, axis=-1), 2, axis=-1))*grad(r.reshape(n,n)) ).flatten())
    mu = np.linalg.norm(r)**2 / ( np.linalg.norm(np.matmul(A, r))**2  + Lambda*tv_term)
    #update
    dTV = dTV - mu*r
    dTV = np.clip(dTV, 0, 255)
    #dTV = 255 * (dTV - np.min(dTV)) / (np.max(dTV) - np.min(dTV))
    
    # energy
    E[i] = 1/2*np.linalg.norm(e.flatten())**2 + Lambda*np.sum(Den.flatten())
    rec[i] = 1/2*np.linalg.norm(e.flatten())**2 
    TV[i] = Lambda*np.sum(Den.flatten())
```

```python
plt.plot(E[1:], label="Total")
plt.plot(rec[1:], label="Reconstruction")
plt.plot(TV[1:], label="TV")
plt.axis('tight')
plt.xlabel('Iteration #')
plt.ylabel('Energy')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.subplot("131")
plt.imshow(approx.reshape(n,n))
plt.title("Norm approximation")
plt.subplot("132")
plt.imshow(dTV.reshape(n,n))
plt.title("TV approximation")
plt.subplot("133")
plt.imshow(cortex)
plt.title("Original")
plt.show()
```

Rescale result:

```python
dTV = 255 * (dTV - np.min(dTV)) / (np.max(dTV) - np.min(dTV))
```

Comparison of reconstruction quality:

```python
print("SNR for Norm reconstruction:", snr(cortex.flatten(), approx))
print("SNR for TV reconstruction:", snr(cortex.flatten(), dTV))
```

## Parameters choice

TV reconstruction is very sensitive to the regularization parameter choice. One way to chosse it can be to look at the SNR of the reconstructed images in regard of the parameter:

```python
def gradient_descent(Lambda, niter=600):

    dTV = np.matmul(A.T, p)

    for i in range(niter):
        # compute the gradient of the euclidean reconstruction error
        e = np.matmul(A, dTV) - p
        grad_rec = np.matmul(A.T, e)
        # compute the gradient of the smoothed TV functional.
        Gr = grad(dTV.reshape(n,n))
        Den = np.sqrt(epsilon**2 + np.sum(Gr**2, axis=2))
        G = -div(Gr / np.repeat(np.expand_dims(Den, axis=-1), 2, axis=-1) )
        # add the two gradients to form the update vector r
        r = grad_rec + Lambda*G.flatten()
        # compute step size according to energy reconstruction values
        #mu = np.matmul(r, grad_rec) / np.linalg.norm(np.matmul(A, r))**2
        tv_term = np.matmul(r, -div( (Gr / np.repeat(np.expand_dims(Den, axis=-1), 2, axis=-1))*grad(r.reshape(n,n)) ).flatten())
        mu = np.linalg.norm(r)**2 / ( np.linalg.norm(np.matmul(A, r))**2  + Lambda*tv_term)
        #update
        dTV = dTV - mu*r
    
    return dTV
```

```python
Lambdas = np.linspace(8e8,9e8,10) #np.linspace(1.8e9,2.1e9,10)
SNR = []

for Lambda in Lambdas:
    # TV reconstruction
    dTV = gradient_descent(Lambda, niter=50)
    # rescale
    dTV = 255 * (dTV - np.min(dTV)) / (np.max(dTV) - np.min(dTV))
    SNR.append( snr(cortex.flatten(), dTV) )
```

```python
plt.plot(Lambdas, SNR)
plt.axis('tight')
plt.xlabel('Lambda')
plt.ylabel('SNR')
plt.show()
```

```python
niter_list = np.linspace(100, 2000,10)
SNR_2 = []

for niter in niter_list:
    # TV reconstruction
    dTV = gradient_descent(Lambda=2.0e9, niter=int(niter))
    # rescale
    dTV = 255 * (dTV - np.min(dTV)) / (np.max(dTV) - np.min(dTV))
    SNR_2.append( snr(cortex.flatten(), dTV) )
```

```python
plt.plot(niter_list, SNR_2)
plt.axis('tight')
plt.xlabel('niter')
plt.ylabel('SNR')
plt.show()
```
