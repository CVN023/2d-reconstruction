# DNN reconstruction

We tackle the ill-posed inverse problem which consists in finding a source $d \in \mathbb{R}^n$ such that $p = Ad$ is verified, where $p \in \mathbb{R}^m$ is an observed projection and $A \in \mathbb{R}^{m \times n}$ represents the projection operator of the imaging system.

Here, we consider an underdetermined problem with $m<n$ ($A$ is a *fat matrix*) which consequently has an infinite number of solutions.

In what follows, $A$ is supposed to be full-rank ($rank(A)=m$).

```python
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as sns
%matplotlib inline
sns.set_context('poster')
sns.set_color_codes()
plot_kwds = {'alpha' : 0.25, 's' : 80, 'linewidths':0}

import scipy as sp
import pylab
import h5py
```

## Data loading:

Projection matrix $A$:

```python
import scipy.io
mat = scipy.io.loadmat('projection_matrices/thinL64_90')
A = mat['H'].todense()
```

```python
A.shape
```

Importing brain dataset:

```python
import h5py
```

```python
brains = []
for i in range(1, 767):
    with  h5py.File('brains-master/%s.mat' %i, 'r') as f:
        brain = np.array(f['cjdata']['image'])
        brain = brain[::8, ::8]
        brains += [brain]

brains = np.array(brains)
print(brains.shape)
```

Computing projections of each 2D image:

```python
projections = []
for i in range(brains.shape[0]):
    p = np.matmul(A, brains[i].flatten())
    p = p.reshape((90,90))
    projections += [p]
    
projections = np.array(projections)
print(projections.shape)
```

```python
fig=plt.figure(figsize=(10, 5))

for i in range(5):
    idx = np.random.randint(766)
    fig.add_subplot(2, 5, i+1)
    plt.imshow(brains[i])
    plt.axis('off')
    fig.add_subplot(2, 5, i+6)
    plt.imshow(projections[i])
    plt.axis('off')
```

Comparison between original image and forward-backward image:

```python
brain_sample = brains[5]
proj = np.matmul(A, brain_sample.flatten())
rec = np.matmul(A.T, proj.T)


fig=plt.figure(figsize=(9, 5))

fig.add_subplot(1, 3, 1)
plt.imshow(brain_sample)
plt.title("Original")
plt.axis('off')

fig.add_subplot(1, 3, 2)
plt.imshow(proj.reshape(90, 90))
plt.title("Projection")
plt.axis('off')

fig.add_subplot(1, 3, 3)
plt.imshow(rec.reshape(64, 64))
plt.title("Image domain")
plt.axis('off')
```

## Learning the Filter-Back Projection

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Conv2D, MaxPooling2D, Flatten, Dropout
```

Model definition:

```python
from keras.datasets import mnist
from keras.layers import Input, Dense, Reshape, Flatten, Dropout, BatchNormalization, UpSampling2D, MaxPooling3D
from keras.layers.advanced_activations import ReLU
from keras.models import Sequential, Model
from keras.optimizers import Adam, RMSprop
from keras import regularizers

plt.switch_backend('agg')

           
            
class FBP(object):
    def __init__(self, width = 64, height= 64, proj_shape=90, nb_proj=90, channels=1, sparse_reg_l1=1e-3, nb_kernels=1):
        self.WIDTH = int(width)
        self.HEIGHT = int(height)
        self.PROJ_SHAPE = int(proj_shape)
        self.NB_PROJ = int(nb_proj)
        self.CHANNELS = int(channels)
        self.SHAPE = (self.WIDTH, self.HEIGHT, self.CHANNELS)
        self.OPTIMIZER = Adam()#lr=0.02, decay=8e-9)
        self.SPARSE_REG_L1 = sparse_reg_l1
        self.NB_KERNELS = nb_kernels
        
        self.fbp = self.filter_back_proj()
        self.fbp.compile(loss='mean_absolute_error', optimizer=self.OPTIMIZER)
        
        
        
    def filter_back_proj(self):
        model = Sequential()
        
        #model.add(Conv2D(self.NB_KERNELS, (1, self.PROJ_SHAPE), activation='linear', use_bias=False, padding='same',
        #                 input_shape=(self.PROJ_SHAPE, self.NB_PROJ, self.CHANNELS)))
        model.add(Conv2D(self.NB_KERNELS, (1, self.PROJ_SHAPE), activation='relu', use_bias=True, padding='same',
                         input_shape=(self.PROJ_SHAPE, self.NB_PROJ, self.CHANNELS)))
        model.add(Conv2D(1, (1, 1), activation='linear', use_bias=False, padding='same'))
        
        model.add(Flatten())
        model.add(Dense(self.WIDTH*self.HEIGHT, use_bias=False, kernel_regularizer=regularizers.l1(self.SPARSE_REG_L1)))
        model.add(Reshape((self.WIDTH, self.HEIGHT, self.CHANNELS)))
        
        return model
    
    

    
    def train(self, X_train, y_train, X_val, y_val, epochs=50, batch = 16, save_interval = 4, print_interval=1, save_weights_int=500):
        self.fbp_losses = []
        self.val_losses = []
        for cnt in range(epochs):

            # train filter_back_proj
            random_index =  np.random.randint(0, len(X_train) - batch)
            y = y_train[random_index : random_index + int(batch)].reshape(int(batch), self.WIDTH, self.HEIGHT, self.CHANNELS)
            proj = X_train[random_index : random_index + int(batch)].reshape(int(batch),
                                                                             self.PROJ_SHAPE, self.NB_PROJ, self.CHANNELS)
            fbp_loss = self.fbp.train_on_batch(proj, y)
            
            
            if cnt % print_interval == 0:
                self.fbp_losses.append(fbp_loss)
                val_loss = self.fbp.evaluate(X_val, y_val, batch_size=batch)
                self.val_losses.append(val_loss)
                print ('epoch: %d, [Reconstruction loss: fbp_loss: %f - Val loss: %f]' % (cnt, fbp_loss, val_loss))
                
            if cnt % save_interval == 0: 
                self.plot_images(X_train=X_train, y_train=y_train, save2file=True, step=cnt)
            
            if cnt % save_weights_int == 0:
                self.fbp.save_weights('10kernels_relu_sparse_reg_dense_1e-3_batch32_%d.h5' % (cnt))
                
                
                
    def plot_images(self, X_train, y_train, save2file=False,  samples=8, step=0):
        filename = "./images/reconstruction_%d.png" % step
        random_index =  np.random.randint(0, len(y_train) - samples)
        y = y_train[random_index : random_index + samples]
        proj = X_train[random_index : random_index + samples].reshape(samples, self.PROJ_SHAPE, self.NB_PROJ, self.CHANNELS)
        predictions = self.fbp.predict(proj)
            
        plt.figure(figsize=(10,10))
        
        for i in range(y.shape[0]):
            plt.subplot(4, 4, i+1)
            pred = predictions[i, :, :, :]
            pred = np.reshape(pred, [ self.HEIGHT, self.WIDTH ])
            plt.imshow(pred)#, cmap='gray')
            plt.axis('off')
            plt.tight_layout()
            #plt.show()
            
            plt.subplot(4, 4, i+9)
            y_i = y[i, :, :, :]
            y_i = np.reshape(y_i, [ self.HEIGHT, self.WIDTH ])
            plt.imshow(y_i)#, cmap='gray')
            plt.axis('off')
            plt.tight_layout()
            #plt.show()
            
        if save2file:
            plt.savefig(filename)
            plt.close('all')
        else:
            plt.show()
```

```python
X_train = projections.copy()
y_train = brains.copy()
```

```python
X_train = np.expand_dims(projections, axis=3)
y_train = np.expand_dims(brains, axis=3)
```

```python
np.random.seed(5)
idx = np.arange(X_train.shape[0])
np.random.shuffle(idx)
X_train = X_train[idx]
y_train = y_train[idx]
```

```python
print(X_train.shape)
print(y_train.shape)
```

Training, Validation and Test splits:

```python
np.random.seed(5)
id_val = int(0.2*X_train.shape[0])
id_test = int(0.4*X_train.shape[0])

X_val = X_train[:id_val]
X_test = X_train[(id_val+1):id_test]
X_train = X_train[(id_test+1):]

y_val = y_train[:id_val]
y_test = y_train[(id_val+1):id_test]
y_train = y_train[(id_test+1):]
```

```python
print(X_val.shape)
print(X_test.shape)
print(X_train.shape)
```

```python
fbp = FBP(sparse_reg_l1=1e-3, nb_kernels=10)
```

```python
fbp.fbp.summary()
```

```python
fbp.train(X_train, y_train, X_val, y_val, epochs=3000, batch = 32, save_interval = 50, print_interval=10)
```

Test:

```python
fbp.fbp.evaluate(X_test, y_test, batch_size=32)
```

```python
def snr(x, y):
    return 20 * np.log10(pylab.norm(x) / pylab.norm(x - y))
```

```python
pred_train = fbp.fbp.predict(X_train)

snr_train = 0
for i in range(len(X_train)):
    snr_train += snr(pred_train[i].flatten(), y_train[i].flatten())
snr_train /= len(X_train)
print("SNR Train set:", snr_train)


pred_val = fbp.fbp.predict(X_val)

snr_val = 0
for i in range(len(X_val)):
    snr_val += snr(pred_val[i].flatten(), y_val[i].flatten())
snr_val /= len(X_val)
print("SNR Validation set:", snr_val)


pred_test = fbp.fbp.predict(X_test)

snr_test = 0
for i in range(len(X_test)):
    snr_test += snr(pred_test[i].flatten(), y_test[i].flatten())
snr_test /= len(X_test)
print("SNR Test set:", snr_test)
```
